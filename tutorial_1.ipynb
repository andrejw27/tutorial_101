# How to add a Fitted Line to a Scatterplot?

## 1. **Fitted Line**
Fitted line is a line that defines the optimal relationship between 2 variables. It is often seen in a regression problem, where we want to draw a regression line given data points of 2 variables that are usually visualize in a *scatterplot*. 

A line is formally defined as:<br>
&emsp;&emsp;y = *m*\*x + *c* <br>
where: <br>
* x and y are the data points
* *m* is the slope of the line 
* *c* is the intercept of the line on y-axis

A fitted line can be achieved by minimizing the distances between data points and the estimated line. There are some ways to add a fitted line to a scatterplot of the data points:
1. Closed-form solution (not covered in this tutorial)
2. Use built-in numpy function: numpy.polyfit
3. Use built-in seaborn function: seaborn.regplot

## 2. **Example Data & Add-On Libraries**
Before we begin to plot, we need to import the required libraries. <br>

```python
import numpy as np #import numpy
import matplotlib.pyplot as plt #import matplotlib to plot the data
import seaborn as sns #import seaborn to create a scatterplot and fitted line
```

Then, we need some data points to be visualized in *scatterplot*

```python
x = np.array([0.05, 0.13, 0.19, 0.24, 0.27, 0.29, 0.32, 0.36, 0.37, 0.40])
y = np.array([0.11, 0.14, 0.17, 0.21, 0.24, 0.32, 0.30, 0.39, 0.42, 0.40])
```

```python
plt.scatter(x, y) #scatterplot the data
```

The figure above shows how the data points look like and how variable x is correlated to variable y. The data points form a kind of linear relationship between x and y but it will become visible, if we draw a fitted line.

## 3. **Example: Add a fitted line to a scatterplot using matplotlib**

This example shows how to draw a fitted line from the data points using matplotlib and numpy. We need a built-in function from numpy, namely polyfit, to help determine the slope *m* and the intercept *c* of the line.

```python
m, c = np.polyfit(x, y, 1) #estimate the slope and the intercept of the line
```

When we call the *polyfit* function, we have to define the data points, which data points belong to x variable and which data belong to y variable, as well as the degree of the fitting line. Since we are interested in a linear line, the degree is set to 1 in the third argument of the function. 

```python
#add linear regression line to scatterplot 
plt.scatter(x, y, color='green') #plot of the original data point
plt.plot(x, m*x+c) #plot of the fitted line
```

In the plot above, we see the original data points are plotted in scatterplot represented in green dots. The fitted line with blue color runs through the data points and creates a line that minimizes the distances between data points to the line itself.

## 4. **Example: Add a fitted line to a scatterplot using seaborn**

This example shows how to draw a fitted line from the data points using seaborn. Seaborn has a built-in function called regplot. This function is able to create a *scatterplot* of data points along with the fitted line automatically. Hence, it does not require us the estimate the slope and the intercept of the line like we do in the previous example using *numpy*.

```python
sns.regplot(x, y, ci=None) #create a scatterplot of the data points along with a fitted line
```

In the *regplot* function, we need to define the data points for x-axis and y-axis in the first two arguments, respectively. In the third argument, there is a parameter *ci*, which stands for confidence interval. If *ci* is set to be any integer number between 0 and 100, then the line will have a translucent band surrounding the fitted line

```python
sns.regplot(x, y, ci=90) #create a scatterplot of the data points along with a fitted line
```
